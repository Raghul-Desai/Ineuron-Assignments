# 1.1 Own Reduce Function

```python
def ownreduce(func, sequence):
  result = sequence[0] 
  for item in sequence[1:]:
    result = func(result, item)
  return result
```

```python
ownreduce(lambda x,y: x*y,[1,2,3,4,5,6,7])
```

# 1.2 Own Filter Function

```python
def ownfilter(func, sequence):
 result = []
 for item in sequence:
  if func(item):
   result.append(item)
 return result
```

```python
ownfilter(lambda x: x>3,[1,2,3,4,5,6,7])
```

# 2 List comprehensions to produce the following Lists

```python
input_list = ['x','y','z']
output = [ item*num for item in input_list for num in range(1,5)  ]
print(str(output))
```

```python
input_list = ['x','y','z']
output = [ item*num for num in range(1,5) for item in input_list  ]
print(str(output))
```

```python
input_list = [2,3,4]
output = [ [item+num] for item in input_list for num in range(0,3)]
print(str(output))
```

```python
input_list = [2,3,4,5]
output = [ [item+num for item in input_list] for num in range(0,4)  ]
print(str(output))
```

```python
input_list=[1,2,3]
output = [ (y,x) for x in input_list for y in input_list]
print(str(output))
```
